```python
# SOURCE: https://stackoverflow.com/questions/10481990/matplotlib-axis-with-two-scales-shared-origin
def align_yaxis(ax1, v1, ax2, v2):
    """adjust ax2 ylimit so that v2 in ax2 is aligned to v1 in ax1"""
    _, y1 = ax1.transData.transform((0, v1))
    _, y2 = ax2.transData.transform((0, v2))
    inv = ax2.transData.inverted()
    _, dy = inv.transform((0, 0)) - inv.transform((0, y1-y2))
    miny, maxy = ax2.get_ylim()
    ax2.set_ylim(miny+dy, maxy+dy)
fig = plt.figure(figsize = (25,8))
ax1 = fig.add_subplot(1, 1, 1)
ax2 = ax1.twinx()
ax1.stem(x_stem, y_stem)
ax2.plot(x_volume, y_volume, 'r-')
align_yaxis(ax1, 0, ax2, 0)
plt.show()
```

<b>We next want to analyse how the behaviour of different stocks are correlated. The correlation is performed on the percentage change of the stock price instead of the stock price.

Load any 5 stocks of your choice into 5 dataframes. Retain only rows for which ‘Series’ column has value ‘EQ’. Create a single dataframe which contains the ‘Closing Price’ of each stock. This dataframe should hence have five columns. Rename each column to the name of the stock that is contained in the column. Create a new dataframe which is a percentage change of the values in the previous dataframe.Using seaborn, analyse the correlation between the percentage changes in the five stocks.</b>

```python
def read_dataset(data):
    df=pd.read_csv(data)
    df=df[df.Series=='EQ']
    df=df.reset_index(drop=True)
    df['Date']=pd.to_datetime(df['Date'])
    df=df.set_index('Date')
    return df
```

```python
df_Maruti=read_dataset('MARUTI.csv')
df_Hdfc=read_dataset('HDFCBANK.csv')
df_Rel=read_dataset('RELIANCE.csv')
df_axis=read_dataset('AXISBANK.csv')
df_tcs=read_dataset('TCS.csv')
```

```python
cp_Maruti=df_Maruti['Close Price']
cp_Hdfc=df_Hdfc['Close Price']
cp_Rel=df_Rel['Close Price']
cp_axis=df_axis['Close Price']
cp_tcs=df_tcs['Close Price']
```

<b>Plotting the 21 day and 34 day Moving average with the average price and decide a Call !

Call should be buy whenever the smaller moving average (21) crosses over longer moving average (34) AND the call should be sell whenever smaller moving average crosses under longer moving average. One of the most widely used technical indicators.<b>

```python
df9=pd.DataFrame()
df9['SMA_21']=data['Close Price'].rolling(21).mean()
df9['LMA_34']=data['Close Price'].rolling(34).mean()
column_names = df9.columns
plt.figure(figsize=(20,8))
plt.plot(df9)
plt.legend(column_names)
plt.show()
```

```python
df9['Close Price']=data['Close Price']
df9
```

```python
p_index       = df9.index[0]
p_row         = df9[:p_index]
p_sma = p_row["SMA_21"][0] > p_row["LMA_34"][0]
df9['Trade_Call'] = 'HODL'

for index, row in df9.iterrows():
    if row["SMA_21"] and row["LMA_34"]:    
        if p_sma and row["SMA_21"] < row["LMA_34"]:
            result = "SELL"
        elif not p_sma and row["SMA_21"] > row["LMA_34"]:
            result = "BUY"
        else:
            result = "HODL" 
        df9.at[index, 'Trade_Call'] = result
        p_sma = row["SMA_21"] > row["LMA_34"]                        
    
df9_Trade_Call = df9[df9['Trade_Call'] != "HODL" ]
df9_Trade_Call
```

```python
fig = plt.figure(figsize=(20,10))
ax1 = fig.add_subplot(111, ylabel="Price")

df9[["Close Price", "SMA_21", "LMA_34"]].plot(ax=ax1)

ax1.plot( df9["Close Price"][ df9['Trade_Call'] == 'BUY' ].index, 
          df9["Close Price"][ df9['Trade_Call'] == 'BUY' ], 
          '^', markersize=15, color='g' )

ax1.plot( df9["Close Price"][ df9['Trade_Call'] == 'SELL' ].index, 
          df9["Close Price"][ df9['Trade_Call'] == 'SELL' ], 
          'v', markersize=15, color='r' )
data["SMA_21"] = df9["SMA_21"] 
data["LMA_34"] = df9["LMA_34"]
data["Trade_Call"] = df9["Trade_Call"]
```

<b>Plot the bollinger bands for this stock - the duration of 14 days and 2 standard deviations away from the average</b>



```python
data=data.reset_index()
data["14_day_SMA"] = data['Close Price'].rolling(14).mean()
data["14_day_STD"] = data['Close Price'].rolling(14).std()
data["14_day_bollinger_upper"] = data["14_day_SMA"] + data["14_day_STD"] * 2
data["14_day_bollinger_lower"] = data["14_day_SMA"] - data["14_day_STD"] * 2
```

```python
plt.figure(figsize=(20,10))
ax = plt.gca()

data.plot(kind='line',x='Date', y='Average Price', ax=ax, color='black')
data.plot(kind='line',x='Date', y='14_day_SMA',    ax=ax, color='blue', linestyle='--')
data.plot(kind='line',x='Date', y='14_day_bollinger_upper',    ax=ax, color='red', linestyle='--')
data.plot(kind='line',x='Date', y='14_day_bollinger_lower',    ax=ax, color='red', linestyle='--')


plt.title("14 day bollinger bands")
plt.ylabel("Price")
plt.show()
```
